```python
df_sorted=df.sort_values(by="Age")
plt.figure(figsize=(10,20))
plt.barh(df_sorted['Player'],df_sorted['Age'],color='skyblue',edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Player')
plt.title('Age of Football Players')
plt.show()
```

```python
sns.set(style="whitegrid")
top10_youngest = df.sort_values('Age').head(10)

plt.figure(figsize=(10, 7))
bars = plt.barh(top10_youngest['Player'], top10_youngest['Age'], color=sns.color_palette("viridis", 10), edgecolor='black')
plt.gca().invert_yaxis()  # Youngest on top

plt.xlabel('Age', fontsize=12,weight='bold')
plt.ylabel('Player', fontsize=12,weight='bold')
plt.title('Top 10 Youngest Football Players', fontsize=16, weight='bold')
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Add data labels at the end of bars
for bar in bars:
    width = bar.get_width()
    plt.text(width + 0.5, bar.get_y() + bar.get_height()/2, f'{int(width)}', va='center', fontsize=11, fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
sns.set(style='whitegrid')
top10_oldest=df.sort_values(by="Age",ascending=False).head(10)
plt.figure(figsize=(10,7))
bars=plt.barh(top10_oldest['Player'],top10_oldest['Age'],color=sns.color_palette('plasma',10),edgecolor='black')
plt.gca().invert_yaxis()
plt.xlabel('Age',fontsize=12,weight='bold')
plt.ylabel('Players',fontsize=12,weight='bold')
plt.title('Top 10 Oldest Football Players',fontsize=16,weight='bold')
plt.grid(axis='x',linestyle='--')

for bar in bars:
  width=bar.get_width()
  plt.text(width+0.5,bar.get_y()+bar.get_height()/2,f'{int(width)}',va='center',fontsize=11,fontweight='bold')

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
league_counts=df['League'].value_counts()
plt.figure(figsize=(10, 6))
sns.countplot(data=df, y='League', hue='League',order=league_counts.index,palette='plasma',edgecolor='black')

plt.title('Player Count by League',fontsize=16,weight='bold')
plt.xlabel('Number of Players',fontsize=12,weight='bold')
plt.ylabel('League',weight='bold',fontsize=12)
plt.tight_layout()
plt.show()

```
